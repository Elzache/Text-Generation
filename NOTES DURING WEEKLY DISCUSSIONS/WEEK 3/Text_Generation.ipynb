```python
import pandas as pd
from sklearn.model_selection import train_test_split
import re
```

```python
df = pd.read_csv('/content/drive/MyDrive/booksummaries.txt', sep='\t', names=['wiki_id', 'freebase_id', 'title', 'author', 'date', 'genres', 'summary'])
df
```

```python
train_test_ratio = 0.9 
#between 0 & 1 (proportion of dataset to include in train split)

train_valid_ratio = 7/9 
#between 0 & 1 (proportion of dataset to include in test split)

#rando
df_full_train, df_test = train_test_split(df, train_size = train_test_ratio, random_state = 1)
df_train, df_valid = train_test_split(df_full_train, train_size = train_valid_ratio, random_state = 1)
```

```python
def build_dataset(df, dest_path):
    f = open(dest_path, 'w')
    data = ''
    summaries = df['summary'].tolist()
    for summary in summaries:
        summary = str(summary).strip()
        summary = re.sub(r"\s", " ", summary)
        bos_token = '<BOS>'
        eos_token = '<EOS>'
        data += bos_token + ' ' + summary + ' ' + eos_token + '\n'
        
    f.write(data)
```

```python
build_dataset(df_train, 'train.txt')
build_dataset(df_valid, 'valid.txt')
build_dataset(df_test, 'test.txt')
```

```python
DATASET="/content"
!fairseq-preprocess \
--only-source \
--trainpref $DATASET/train.txt \
--validpref $DATASET/valid.txt \
--testpref $DATASET/test.txt \
--destdir data-bin/summary \ 
--workers 20

#the preprocessed data will be saved in the directory specified by the --destdir
```

```python
!CUDA_VISIBLE_DEVICES=0 fairseq-train --task language_modeling \ #GPU to use as the 0th (CUDA_VISIBLE_DEVICES), task as language modeling (--task)
data-bin/summary \
--save-dir checkpoints/transformer_summary \
--arch transformer_lm --share-decoder-input-output-embed \
--dropout 0.1 \
--optimizer adam --adam-betas '(0.9, 0.98)' --weight-decay 0.01 --clip-norm 0.0 \
--lr 0.0005 --lr-scheduler inverse_sqrt --warmup-updates 4000 --warmup-init-lr 1e-07 \
--tokens-per-sample 512 --sample-break-mode none \
--max-tokens 2048 --update-freq 16 \
--fp16 \
--max-update 50000 \
--max-epoch 5
```

```python
!fairseq-eval-lm data-bin/summary \ #Evaluate the perplexity of a trained language model.
--path checkpoints/transformer_summary/checkpoint_best.pt \
--max-sentences 2 \
--tokens-per-sample 512 \
--context-window 400
```
